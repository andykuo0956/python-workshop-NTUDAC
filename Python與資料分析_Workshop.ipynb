### List（串列）

List 是 Python 中最基本且重要的資料結構，List 中包含的元素是**不需要具有相同的資料類型**的！

同時也能夠**隨意改變** List 中的元素，甚至能進行新增、刪減、排序等動作。

* List 舉例

```python
number = [1, 2, 3, 4, 5]
name = ['Amy', 'Tracy', 'John', 'Betty', 'Adam']
day_of_week = ['Monday', 1, 'Tuesday', 2, 'Wednesday', 3, 'Thursday', 4, 'Friday', 5]
```

* 如何建立一個不包含任何元素的 List 呢？

```python
# 方法一
list_1 = []
print(list_1)
```

```python
# 方法二
list_2 = list()
print(list_2)
```

Python 中，List 的索引（index）由 0 作為起始，因此在回傳對應索引的值時，要記得是從 0 開始而非從 1 開始！

* 要怎麼在 List 找到第一個值呢？

```python
name = ['Amy', 'Tracy', 'John', 'Betty', 'Adam']
name
```

```python
print("在 name 中的第一個值為：", name[0])
```

* 如果想要找到最後一個值呢？

```python
print("在 name 中的最後一個值為：", name[-1])
```

* 如果想要第二個到第四個值呢？

```python
print("在 name 中的第二個到第四個的值為：", name[1:4])
```

有注意到冒號嗎？這個冒號的意思我們可以想像成刪節號（⋯⋯）的概念！

只要輸入起始位置的 index 就能回傳對應的值了！

**注意！**在冒號右側的 index 值，是不會回傳該 index 本身對應到的值，而是 index - 1 對應到的值，使用上需要特別小心！

* 想知道 John 屬於哪個 index

```python
# index(obj)
print("John 的 index 為: ", name.index("John"))
print(name)
```

### Tuple（元組）

Python 中的元組與串列相似，最大的不同為**元組內的元素不得進行修改！**

* Tuple 舉例

```python
number = (1, 2, 3, 4, 5)
name = 'Amy', 'Tracy', 'John', 'Betty', 'Adam'
day_of_week = ('Monday', 1, 'Tuesday', 2, 'Wednesday', 3, 'Thursday', 4, 'Friday', 5)
```

* 如何建立一個不包含任何元素的 Tuple 呢？



```python
# 方法一
tuple_1 = ()
print(tuple_1)
```

```python
# 方法二
tuple_2 = tuple()
print(tuple_2)
```

在 Tuple 中，尋找 index 對應下之值的方法跟 List 一模一樣！

* 想要找第二個到第四個值

```python
print("在 name 中的第二個到第四個的值為：", name[1:4])
```

不得修改的特性意即不得於該元組中新增或刪除任何「元素」，必須是以「元組」作為單位進行新增或是刪除

也因此雖然 Tuple 中的值是不允許修改的，但可以將元組進行串接！

* 非法的 Tuple 修改

```python
tuple_1 = (1, 2, 3, 4, 5)
tuple_1[0] = 0 #不能這樣修改
```

* Tuple 之間的連結

```python
tuple_2 = (6, 7, 8, 9, 10)
tuple_3 = tuple_1 + tuple_2
print("連結後的新元組：", tuple_3)
```

若要刪除的話只能刪除元組！

```python
# del
print("原本的元組：", tuple_3)
del tuple_3
print("刪除後的元組：", tuple_3) #無法印出，因為已經被刪除了
```

### Dictionary（字典）

在 Python 中，Dictionary 是一種可變的資料結構，可以儲存任意資料類型。

其中以 key:value 作為整體架構的主軸。key 與 value 中間用冒號 (:) 進行分割；每一組 key & value 之間則是以逗號 (,) 進行分割。

架構如下：

dictionary = {key1: value1, key2: value2}

* 如何建立一個 Dictionary 呢？

```python
# 方法一
dict_1 = dict()
dict_1['one'] = 1
dict_1['two'] = 2
dict_1['three'] = 3
print(dict_1)
```

```python
# 方法二
dict_2 = {'one': 1,
          'two': 2,
          'three': 3}
print(dict_2)
```

* 想知道 Dictionary 中會有哪些 key

```python
student = {'Name': 'Tracy', 'Birthday': '1999/09/09', 'Score': 80}
print(student.keys())
```

* 想知道 Dictionay 中會有哪些 value

```python
print(student.values())
```

* 想得到 Student 中的 Name 是什麼

```python
# 方法一
print(student.get('Name'))
```

```python
# 方法二
print(student['Name'])
```

```python
# 方法一與方法二差異
print("方法一：", student.get('Number'))
print("方法二：", student['Number'])
```

有發現嗎？

如果我在一個字典裡面直接尋找本來就不在裡面的 key 值，這時候就會報 KeyyError 的錯誤

但是如果用 .get 尋找，就不會有報錯的問題，只會回傳空值，因此會建議使用 .get 來尋找 key 對應底下的值唷！

## 1-2｜迴圈

### While Loop

當你想要檢視的條件成立時，就做你想要做的事；每次做完就重新檢視一次，直到條件不成立為止。



* 想知道 0 到 100 的加總為何

```python
sum = 0
i = 1
while i <= 100:
  sum = sum + i
  i = i + 1

print(sum)
```

遇到無限迴圈怎麼辦？這時候需要使用 continue 與 break 的概念！

* break: 停止動作並跳出當前所在的 loop scope

* continue: 停止動作並執行所在的 loop scope 的下一圈

```python
n = 5

while 1:
  n += 2
  print("現在的 n 為：", n)

  if n % 19 == 0:
    break

  if n % 3 == 0:
    continue
```

### For Loop

* 單層 For Loop

在迴圈中可以適時善用 range 跟 len 的特性！

range 的基本介紹與用法：
  * range(n)
  * range(start, end)
  * range(start, end, step)

* range(n)
* 想成一個長度為 n 的 list

```python
# range(n)
n = 10

for i in range(n):
  print(i)
```

* range(start, end)
* 最後一個東西會是 end-1
* range(end) = range(0, end)

```python
# range(start, end)
start = 0
end = 10

for i in range(start, end):
  print(i)
```

* range(start, end, step)
* 多了一個 step 去設定每次要增加的數量
* range(end) = range(0, end) = range(0, end, 1)

```python
start = 0
end = 10
step = 2

for i in range(start, end, step):
  print(i)
```

如果要把 List 中的元素全部 print 出來有哪些方法啊？

* 用 range & len

```python
# 方法二
animals = ['cat', 'dog', 'mouse', 'rabbit', 'deer', 'dolphin', 'shark', 'crocodile', 'hedgehog']

for i in range(len(animals)):
  print(animals[i])
```

## 1-3｜函數

### Function

Function 在 Python 中為一個被組織好的函數，其性質是可以重複進行使用，就不用一一的打程式碼，超級方便！


* 定義一個函數

```python
def BMI(height_cm, weight):
  height_m = height_cm / 100
  bmi = weight / (height_m**2)
  return bmi
```

```python
# John 的身高為 180 公分；體重為 75 公斤
john_bmi = BMI(180, 75)
print("John 的 BMI 為：", john_bmi)
```

```python
student = pd.DataFrame(
    {
      'Name':   ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'N', 'O'],
      'Gender':    ['M', 'M', 'M', 'F', 'F', 'M', 'F', 'M', 'F', 'M', 'M', 'M', 'F', 'M', 'M'],
      'Age':    [38, 28, 31, 34, 28, 28, 36, 33, 22, 39, 22, 24, 31, 29, 22],
      'Height': [1.74, 1.51, 1.67, 1.87, 1.8, 1.51, 1.85, 1.89, 1.81, 1.72, 1.75, 1.64, 1.9, 1.62, 1.61],
      'Weight': [45, 63, 39, 45, 67, 66, 53, 45, 72, 46, 58, 44, 73, 70, 51],
    }
)
student
```

```python
def BMI_1(r):
  return round(r['Weight'] / (r['Height'] ** 2), ndigits=2)
```

```python
student['BMI_apply1'] = student.apply(BMI_1, axis=1)

student
```

* 利用 lambda 建立新欄位

```python
def BMI_2(weight, height):
  return round(weight / (height ** 2), ndigits=2)
```

```python
student['BMI_apply2'] = student.apply(lambda r: BMI_2(r['Weight'],r['Height']), axis=1)

student
```

## 2-6｜迴圈在 pandas 中可以扮演的角色

介紹了那麼多 pandas 常見的語法，但還是想不到迴圈可以用在哪個地方⋯⋯

分享一個你可能會在 pandas 裡面使用到迴圈的情境！！！

我們延續剛剛已經得到 BMI 值的 student 表

* 想要觀察 BMI_apply1 中的 BMI 值，並修改 BMI_apply2 裡面的值
* 舉例：BMI_apply1 為 20.68 時，此時對應的 BMI_apply2 的值會變成 "健康體重"
* BMI 指標參考：

> 低於 18.5：體重過輕

> 18.5-24：健康體重

> 24~27：體重過重

> 27 以上：肥胖

```python
for i in range(len(student)):

  if student.loc[i, 'BMI_apply1'] < 18.5:
    student.loc[i, 'BMI_apply2'] = '體重過輕'
  
  elif student.loc[i, 'BMI_apply1'] >= 18.5 and student.loc[i, 'BMI_apply1'] < 24:
    student.loc[i, 'BMI_apply2'] = '健康體重'

  elif student.loc[i, 'BMI_apply1'] >= 24 and student.loc[i, 'BMI_apply1'] < 27:
    student.loc[i, 'BMI_apply2'] = '體重過重'
  
  else:
    student.loc[i, 'BMI_apply2'] = '肥胖'
```
